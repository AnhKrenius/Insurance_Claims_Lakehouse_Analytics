```python
adls_account = "insurancedatalake01"
storage_key = os.environ.get("AZURE_STORAGE_KEY")

spark.conf.set(
    f"fs.azure.account.key.{adls_account}.dfs.core.windows.net",
    storage_key
)
```

```python
adls_account = "insurancedatalake01"
container = "datalake"
raw_path = f"abfss://{container}@{adls_account}.dfs.core.windows.net/raw"
bronze_path = f"abfss://{container}@{adls_account}.dfs.core.windows.net/bronze"
```

```python
df_customer = (
    spark.read.options(header=True, inferSchema = True)
    .csv(f"{raw_path}/customers/*.csv")
)
display(df_customer)
```

```python
df_policy = (
    spark.read.options(header = True, inferSchema = True)
    .csv(f"{raw_path}/policies/*.csv")
)
df_policy.display()
```

```python
df_handler = (
    spark.read.options(header = True, inferSchema = True)
    .csv(f"{raw_path}/handlers/*.csv")
)
df_handler.display()
```

```python
df_claims = (
    spark.read.options(header = True, inferSchema = True)
    .csv(f"{raw_path}/claims/*.csv")
)
display(df_claims)
```

```python
tables = {
    "customer": df_customer,
    "claims":df_claims,
    "policy": df_policy,
    "handler": df_handler
}
for name, df in tables.items():
    df.write.format("delta").mode("overwrite").save(f"{bronze_path}/{name}")
```
